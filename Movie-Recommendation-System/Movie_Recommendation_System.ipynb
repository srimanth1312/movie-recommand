```python
#Importing Libraries
import pandas as pd
import numpy as np
```

```python
#Importing DataSet
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

Data Preprocessing

```python
movies.head()
```

```python
credits.head()
```

```python
#Merging DataFrames
movies=movies.merge(credits,on='title')
movies.info()
```

```python
movies.head()
```

```python
#Features used : genres,movie_id,keywords,overview,cast,crew

movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.iloc[0].genres
```

```python
import ast
def get_value(obj):
    l=[]
    for i in ast.literal_eval(obj):  #we are converting string of lists into lists using literal_eval
        l.append(i['name'])
    return l
```

```python
movies['genres']=movies['genres'].apply(get_value)
```

```python
movies['keywords']=movies['keywords'].apply(get_value)
```

```python
movies.head()
```

```python
def get_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if(i['job']=='Director'):
            l.append(i['name'])
            break
    return l
        
```

```python
movies['crew']=movies['crew'].apply(get_director)
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
def convert(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies.tail()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies_t=movies[['movie_id','title','tags']]
```

```python
movies_t.head()
movies_t['title']
```

```python
movies_t['tags']=movies_t['tags'].apply(lambda x:" ".join(x))
```

```python
movies_t['tags'][1]
```

```python
import nltk
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
def stem(text):
    l=[]
    for i in text.split():
        l.append(stemmer.stem(i))
    s = " ".join(l)
    return s
```

```python
movies_t['tags'].apply(stem)
movies_t.head()
```

Implementing Model 

```python
#text vectorization(bag of words method)
#stop words => words which are used for sentence formation but has no meaning

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(movies_t['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
#Finding cosine distance (angle between vectors)
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity.shape
#similarity has the cosine distance between every movie with all other movies
```

```python
def get_movies(movie):
    index = movies_t[movies_t['title']==movie].index[0]
    distances = list(enumerate(similarity[index]))
    distances.sort(reverse=True,key=lambda x:x[1])
    l=distances[1:6]
    for i in l:
        print(movies_t['title'][i[0]])
```

```python
get_movies("Pirates of the Caribbean: At World's End")
```
